# **Movie Data Portfolio**



 Step 1: **Understand the Dataset**
    movies.csv contains:


*   Movie names, genres, scores, votes
*   Directors, writers, stars


*   Budgets, gross earnings, companies
*   Release year, country, runtime










```python
import pandas as pd

# Load the CSV file
movies_df = pd.read_csv("/movies.csv")
```

Step 2: **Clean the Data**
      Why: Missing values can mislead visuals

```python
cleaned_movies = movies_df.dropna(subset=['score', 'votes', 'budget', 'gross', 'genre', 'year'])
```

**1. Top 10 Genres by Average IMDB Score**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the visual style
sns.set(style="whitegrid")

# Calculate top 10 genres by average IMDB score
top_genres = cleaned_movies.groupby('genre')['score'].mean().sort_values(ascending=False).head(10)

# Create barplot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title("Top 10 Genres by Average IMDB Score")
plt.xlabel("Average IMDB Score")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

**2. Budget vs Gross Revenue**

```python
sns.scatterplot(data=cleaned_movies, x='budget', y='gross', alpha=0.5)
sns.regplot(data=cleaned_movies, x='budget', y='gross', scatter=False, color='red')
```

**3. Number of Movies Released Per Year**

```python
sns.histplot(cleaned_movies['year'], bins=40, kde=False, color="skyblue")
```

**4. Correlation matrix of Movie Features**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix (numeric columns only)
correlation_matrix = cleaned_movies.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Movie Features")
plt.tight_layout()
plt.show()
```
